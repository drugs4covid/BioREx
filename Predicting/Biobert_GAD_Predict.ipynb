```python
import transformers
```

```python
from transformers import DistilBertForTokenClassification, DistilBertConfig, AutoModelForTokenClassification, BertForSequenceClassification,AutoModelForSequenceClassification
```

```python
model= AutoModelForSequenceClassification.from_pretrained("ChrisUPM/BioBERT_Re_trained")
```

```python
from transformers import AutoTokenizer
```

```python
tokenizer= AutoTokenizer.from_pretrained("ChrisUPM/BioBERT_Re_trained")
```

```python
from transformers import pipeline

generator = pipeline(task="text-classification", model=model, tokenizer=tokenizer)
```

```python
generator("These results suggest that the C1772T polymorphism in @GENE$ is not involved in progression or metastasis of @DISEASE$.")
```
